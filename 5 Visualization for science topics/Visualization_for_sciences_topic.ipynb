## Heat Equation

The heat equation is an important partial differential equation (PDE) which describes the distribution of heat (or variation in temperature) in a given 
region over time. For better understanding, it is very important that we understand the difference between heat and temperature. 

Heat is a process of energy transfer as a result of temperature difference between the two points. Thus, the term 'heat' is used to describe the energy transferred through the heating process. Temperature, on the other hand, is a physical property of matter that describes the hotness or coldness of an object or environment. Therefore, no heat would be exchanged between bodies of the same temperature.

The heat equation is a parabolic partial differential equation that describes the distribution of heat (or variation in temperature) in a given region over time. For a function (x,y,z,t) of three spatial variables (x,y,z) (see Cartesian coordinates) and the time variable t, the heat equation is:

\begin{equation*}
    \frac{\partial u}{\partial t} - \alpha \left(\frac{\partial^2 u}{\partial x^2} + \frac{\partial^2 u}{\partial y^2} + \frac{\partial^2 u}{\partial z^2} \right) = 0 
\end{equation*}

More generally in any coordinate system : 

\begin{equation*}
\frac{\partial u}{\partial t} - \alpha \nabla^2 u = 0 
\end{equation*}

Where α is a positive constant, and Δ or ∇$^2$  denotes the Laplace operator. In the physical problem of temperature variation, u(x,y,z,t) is the temperature and α is the thermal diffusivity. For the mathematical treatment it is sufficient to consider the case α = 1.

### Heat Equation in 1D

The heat equation in one dimension is written as the following:

\begin{equation*}
\frac{\partial u}{\partial t} = \alpha \left( \frac{\partial^2 u}{\partial x^2} \right)
\end{equation*}

where U (x; t) is a function of temperature. 
In this case we can think of a one dimensional rectangular thin wire with length x.
Ignore the width and height dimensionality. The one end of the length of the wire is set at 0$^{\circ}$  whereas the other end is set at 0$^{\circ}$ C. These are its boundary conditions. We also need to
specify the temperature at every position on the wire at time; t$_0$; (the initial conditions). To 
4
solve this one-dimensional heat problem, we need to transform the above heat equation into
an explicit method using the second-order central Finite Difference Method.

Therefore,
explicitly we can write the one dimensional heat equation as:

\begin{equation*}
\frac{u_j^{n+1} - u_j^n}{\Delta t} = \alpha \left( \frac{u_{j+1,n} - 2u_{j,n} + u_{j-1,n}}{\Delta x^2} \right)
\end{equation*}

The value should be much smaller than 1, other-wise you get unexpected errors.
In this case, we assume that is a constant although later we are going to see that alpha could
be de ned as a function that depends on space. As we can see, the heat equation in 1-D
explicit form is straight forward because the right hand side has only one term. 


### Heat Equation in 2D

The heat equation in two dimension is written as the following:

\begin{equation*}
\frac{\partial u}{\partial t} = \alpha \left( \frac{\partial^2 u}{\partial x^2} + \frac{\partial^2 u}{\partial y^2} \right)
\end{equation*}

where u is the temperature at position (x, y) and time t, moreover D
is conductivity equation.

Before we elaborates numerical method for transport equation, then let’s
consider the following discrete properties:
* Assume that our spatial domain is in [0, Lx ]×[Ly ]. If spatial domain is divided by Nx and Ny points, then spatial space for x and y direction
are given by ∆x = Lx/Nx and ∆y = Ly /Ny. Thus xi = i × ∆x and yi = i × ∆y are the points at index i where i ∈ {0, 1, · · · , N}.
* Assume that our time domain is in [0,Tfin]. Similar to spatial domain,
if time domain is divided by ∆t = Tfin/(T − 1) time step, then
$t^n$ = n × ∆t with n ∈ {0, 1, · · · ,T} is time at point n.

Then using finite difference method from Taylor expansion, we have:

\begin{align}
&\frac{u^{n+1}_{i,j} - u^{n}_{i,j}}{\Delta t} = \mu \left( \frac{u^n_{i+1,j} -2u^n_{i,j} + u^n_{i-1,j}}{\Delta x^2} + \frac{u^{n}_{i,j+1} - 2u^{n}_{i,j} + u^{n}_{i,j-1}}{\Delta y^2} \right),\\
&u_{i,j}^0=f(x_i,x_j,0), \\
&\{u_{i,j}^n\}_{j=0,M_y} =0,  \\
&\{u_{i,j}^n\}_{i=0,M_x} =0, 
\end{align}

### 2D Visualization

Please solve the heat equation

\begin{equation*}
    \frac{\partial u}{\partial t} = D \left(\frac{\partial^2 u}{\partial x^2} + \frac{\partial^2 u}{\partial y^2} \right) 
\end{equation*}

and the initial density of vehicle on domain $[0,Lx]\times[0,Ly]$,

\begin{equation*}
    u(x,y,0)=\nu(x,y)= \max(0, -10(x^2 +y^2) + 50)
\end{equation*}

Moreover, let use $N_x = N_y =50$, $T=101$, $D=1$, $L_x=L_y=10$. And please use Dirichlet boundary condition for boundary. 

Plot the plot at final time $Tfin= 0.25$, $Tfin= 0.5$ and $Tfin= 1$


The initial condition can be seen in the following Figure
<center><img src = "https://imgur.com/5TdIwcI.png" align = "center"> </center>

Setup and initial condition

```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt


def nu(x, y):
    return max(0,(-10*((x**2)+(y**2)))+50)

Lx = 10
Ly = 10

Nx = 50
Ny = 50

D = 1

x = np.linspace(-5,5,Nx) #for plot purpose
y = np.linspace(-5,5,Ny) #for plot purpose
X,Y = np.meshgrid(x,y)
dx =  Lx / Nx
dy = Ly / Ny
print("dx = ",dx)
print("dy = ",dy)

T =  101
Tfin =  0.5 
dt = Tfin/(T-1)
print("dt = ",dt)

#Initial condition
u = np.zeros((Nx,Ny))
u_n = np.zeros((Nx,Ny))

#setting initial condition u
for i in range(0,Nx):
  for j in range(0,Ny):
    u[i][j] = nu(x[i],y[j])

#print(rho)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, u, rstride=1, cstride=1, cmap='viridis', edgecolor='none')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.show()

cp = plt.contourf(X, Y, u)
plt.colorbar(cp)
plt.show()
```

Loop process

```python
time =0 
while(time <=Tfin):
  time = time + dt
  for i in range(0,Nx-1):
    for j in range(0,Ny-1):
      u_n[i][j] = u[i][j] + (D*dt/dx**2)*((u[i+1][j]-2*u[i][j]+u[i-1][j]) + (D*dt/dy**2)*(u[i][j+1]-2*u[i][j]+u[i][j-1]))
   
  #boundary
  for i in range(0,Nx):
     u_n[i][0] = 0
     u_n[i][Ny-1] = 0
  for j in range(0,Ny):
     u_n[0][j] = 0
     u_n[Nx-1][j] = 0
  u = u_n

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, u, rstride=1, cstride=1, cmap='viridis', edgecolor='none')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set(xlim=(-5, 5), ylim=(-5, 5), zlim=(0,50))
plt.show()
plt.savefig("aa.png")

cp = plt.contourf(X, Y, u)
plt.colorbar(cp)
plt.show()
```

### Heat Equation in 3D

The heat equation in two dimension is written as the following:

\begin{equation*}
\frac{\partial u}{\partial t} = \alpha \left( \frac{\partial^2 u}{\partial x^2} + \frac{\partial^2 u}{\partial y^2} +  \frac{\partial^2 u}{\partial z^2} \right)
\end{equation*}

where u is the temperature at position (x, y) and time t, moreover D
is conductivity equation. 

Then using finite difference method from Taylor expansion, we have:

\begin{align}
&\frac{u^{n+1}_{i,j} - u^{n}_{i,j}}{\Delta t} = \mu \left( \frac{u^n_{i+1,j} -2u^n_{i,j} + u^n_{i-1,j}}{\Delta x^2} + \frac{u^{n}_{i,j+1} - 2u^{n}_{i,j} + u^{n}_{i,j-1}}{\Delta y^2} + \frac{u^{n}_{i,j+1} - 2u^{n}_{i,j} + u^{n}_{i,j-1}}{\Delta z^2} \right)
\end{align}

### Rendering Collection of Figures to be Video



We shall go through two examples. The first one reads images from the file system and creates a video. The second example creates a video directly from the programmatically generated numpy arrays.

To create a video from Image Arrays, follow the below sequence of steps.

1. Initialize a Video Writer with the following items specified.
  * Output Video File Name
  * fourcc code that specifies the codec
  * Number of Frames per Second
  * Video Frame Size
2. Write each image array to the Video Writer object.
3. Release the Video Writer.


Example 1: Create Video from Images in a Folder

In this example, we shall read images, present in a folder, one by one. Then we shall use Video Writer to write each image, in a loop, to the video output file.

```python
img_array = []
for filename in glob.glob('gdrive/My Drive/Colab Notebooks/Images/*.png'):
    img = cv2.imread(filename)
    print(img.shape)
    height, width, layers = img.shape
    size = (width,height)
    img_array.append(img)


out = cv2.VideoWriter('project.avi',cv2.VideoWriter_fourcc(*'DIVX'), 1, size)

for i in range(len(img_array)):
    out.write(img_array[i])
out.release()
```

* 'output_video.avi' is the video output file name.
* cv2.VideoWriter_fourcc(*'DIVX') is the codec.
* 60 is the number of frames per second. So, 60 images shall be used to create a video of duration one second. You may change this value as per the requirement.
*frameSize = (500, 500) defines the width and height of the output video.

In the above program, the for loop reads all the .jpg files one by one to a numpy array using cv2.imread(). And the image numpy array is written to video file using Video Writer.

Please note that the order of images that is written to the video file depends on how glob reads the files from the folder.

You may control the order of files you read in the for loop, thus controlling the order of images in the video.
